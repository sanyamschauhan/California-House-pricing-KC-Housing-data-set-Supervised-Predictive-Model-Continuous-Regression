```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
housing1 = pd.read_csv('/Users/sanyamsinghchauhan/Desktop/Data/Python/KC_House.csv')
```

```python
housing1.head(10)
```

```python
housing1.info()
```

```python
housing_new = housing1.iloc[:,[2,3,4,5,7,8,9,11,12,13,17]]
```

```python
housing_new.info()
```

```python
sns.heatmap(housing_new.corr(),annot=True,cmap='viridis')
```

```python
housing = housing_new.dropna()
```

```python
housing.info()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0,10))
scaler.fit_transform(housing)
```

```python
housing.head(10)
```

```python
target = housing.iloc[:,0:1]
features = housing.iloc[:,2:]
```

```python
from sklearn.model_selection import train_test_split
features_train,features_test,target_train,target_test= train_test_split(features,target, test_size=0.20, random_state=10)
```

# Linear Regession Model

```python
from sklearn.linear_model import LinearRegression
model_linear= LinearRegression()
model_linear.fit(features_train,target_train)
```

```python
t_linear_predict = model_linear.predict(features_test)
t_linear_predict
```

```python
from sklearn.metrics import mean_squared_error, r2_score, roc_curve, auc
mse_linear = mean_squared_error(target_test,t_linear_predict)
mse_linear
r2_linear = r2_score(target_test,t_linear_predict)
r2_linear
```

```python
plt.scatter(target_test,t_linear_predict)
plt.title('Predictions w.r.t actual value')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
```

# Support Vector Machine

```python
from sklearn.svm import SVR
model_svm = SVR(kernel='linear')
model_svm.fit(features_train,target_train)
```

```python
t_svm_predict = model_svm.predict(features_test)
t_svm_predict
```

```python
mse_svm= mean_squared_error(target_test,t_svm_predict)
mse_svm
r2_svm=r2_score(target_test,t_svm_predict)
r2_svm
```

```python
plt.scatter(target_test,t_svm_predict)
plt.title('Predictions w.r.t actual value')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor
model_dt = DecisionTreeRegressor(random_state=20)
model_dt.fit(features_train,target_train)
```

```python
t_dt_predict = model_dt.predict(features_test)
t_dt_predict
```

```python
mse_dt= mean_squared_error(target_test,t_dt_predict)
mse_dt
r2_dt=r2_score(target_test,t_dt_predict)
r2_dt
```

```python
plt.scatter(target_test,t_dt_predict)
plt.title('Predictions w.r.t actual value')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
```

# Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
model_rf = RandomForestRegressor(n_estimators=500,random_state=10)
model_rf.fit(features_train,target_train)
```

```python
target_rf_predict = model_rf.predict(features_test)
target_rf_predict
```

```python
mse_rf = mean_squared_error(target_test,target_rf_predict)
r2_rf= r2_score(target_test,target_rf_predict)
r2_rf
```

```python
plt.scatter(target_test, target_rf_predict)
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.title("Linear Regression Model - Actual vs Predicted")
```
